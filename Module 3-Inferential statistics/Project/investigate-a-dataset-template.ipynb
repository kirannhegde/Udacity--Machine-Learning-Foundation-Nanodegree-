## Data Wrangling

For my exploratory questions, the only data wrangling that needs to be done is to remove columns from the dataframe, which are of not being used in the analysis. 

### General Properties

```python
# Load your data and print out a few lines. Perform operations to inspect data
#types and look for instances of missing or possibly errant data.

#First step is to load the csv into a dataframe
filename = "./tmdb-movies.csv"
tmdb_df = pd.read_csv(filename)
```

```python
#Print the top 10 rows of the loaded dataframe
print("Printing the top 10 rows of the loaded dataframe...")
tmdb_df.head(10)
```

```python
#Print the bottom 10 rows of the loaded dataframe
print("Printing the bottom 10 rows of the loaded dataframe...")
tmdb_df.tail(10)
```

### Observation from bottom 10 rows:
Printing the bottom 10 rows of the dataframe reveals that there are entries in the budget and revenue columns with values of 0. For any meaningful analysis related to budget and revenues, we need to trim the rows where the budget and revenue columns have 0 values.

```python
#Inspect the datatype of all the columns
tmdb_df.dtypes
```

### Columns of interest in the dataframe:
For this analysis, the only columns of interest are the budget, revenue and release_year columns in the loaded dataframe. However, i will retain few more columns in addition to these so as to explore more questions in the foreseaable future.

```python
#Drop all columns other than the columns of interest
cols_to_drop = ['id','homepage','tagline','keywords','overview','genres','budget_adj','revenue_adj']
tmdb_df.drop(cols_to_drop,axis=1,inplace=True)


```

```python
#The dataframe with dropped columsn looks like below
tmdb_df
```

### Data Cleaning
The data cleaning process consists of the following steps:

i)Drop duplicate rows, if any

ii)Drop rows which contain values less than 0 in either of budget, revenue and runtime column

####  Reason for filtering the data in the budget and revenue column
In this new dataframe with reduced number of columns, we now need to filter rows which contain 0 in either the budget or the revenue columns.  We are doing this as one of the exploratory questions deals about determining if higher budget movies translate into higher revenues.

```python
#Get the number of rows in the dataframe before any cleanup is done. 
len_before_cleanup = len(tmdb_df.index)
print("Total number of records before data cleanup is:{}".format(len_before_cleanup))
```

```python
#Start inspection of the data.
#First step is to make sure that there are no duplicates in the csv file.
tmdb_df.drop_duplicates(keep = 'first', inplace = True)
len_after_cleanup = len(tmdb_df.index)
print("Total number of records after cleanup is:{}".format(len_after_cleanup))
```

#### **Observation after data cleanup**:
It seems like one duplicate row has been cleaned after the data cleanup operation.

```python
#Filter rows where either the budget,revenue or the runtime  is not greater than 0
#This step ensures that the dataframe contains only rows with valid values for budget, revenue and runtime. 
tmdb_df = tmdb_df[(tmdb_df.budget > 0) & (tmdb_df.revenue > 0) & (tmdb_df.runtime > 0)]


```

```python
#Inspect the number of row remaining after the data cleanup operation.
print("Total number of records in the dataframe after the final data cleanup operation is:{}".format(len(tmdb_df)))
tmdb_df
```

## Exploratory Data Analysis



### Research Question 1 (Which movie has the highest budget?)

```python
#In this cell, we will find out the movie with the highest budget
index = tmdb_df['budget'].idxmax(axis='rows')
tmdb_df.loc[index]
```

### Research Question 2  (Which movie raked in the highest revenues?)

```python
#In this cell, we will find out the movie with the highest revenue
index = tmdb_df['revenue'].idxmax(axis='rows')
tmdb_df.loc[index]
```

### Research Question 3 (How have the revenue collections fared compared to the budget?)

```python
plt.xlabel('Budget')
plt.ylabel('Revenues')
plt.scatter(tmdb_df['budget'],tmdb_df['revenue'])
```

### Research Question 4 (What is the average runtime of the movies?)

```python
avg_runtime = tmdb_df['runtime'].mean()
print("The average runtime of the movies is:{}".format(avg_runtime))
```

```python
#Plot a histogram of the movie runtimes in the filtered dataset
plt.xlabel('Movie runtime in minutes')
plt.ylabel('Number of movies')
plt.title('Movie runtime distribution')
plt.hist(tmdb_df['runtime'])

```

### Research Question 5 (Is the popularity of a movie related to its runtime?)

```python
#Plot runtime on the x-axis and popularity on the y-axis
#Can you infer anything from the scatter plot?
plt.xlabel('Movie runtime in minutes')
plt.ylabel('Movie popularity')

plt.scatter(tmdb_df['runtime'],tmdb_df['popularity'])
```

### Research Question 5 (Which year has the maximum number of movies released? Plot a histogram)

```python
plt.xlabel('Year of release')
plt.ylabel('Number of movies')
plt.hist(tmdb_df['release_year'])
```
